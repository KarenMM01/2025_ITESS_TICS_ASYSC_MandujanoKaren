## Primer Ejercicio 

```python
import numpy as np
import matplotlib.pyplot as plt

V = 1
T = 2*np.pi           
w = 2*np.pi / T    

t = np.linspace(0, 2*T, 2000)

v = np.zeros_like(t)

for i in range(len(t)):
    tp = t[i] % T
    if 0 <= tp <= T/2:
        v[i] = V * np.sin(w * tp)
    else:
        v[i] = 0

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.plot(t, v)
plt.title("Primer Ejercicio")
plt.xlabel("t")
plt.ylabel("v(t)")
plt.grid(True)
plt.show()
```

## Segundo Ejercicio

```python
import numpy as np
import matplotlib.pyplot as plt


V = 1
T = 2*np.pi
w = 2*np.pi / T


t = np.linspace(0, 2*T, 2000)

N = 20  
v_sf = 2*V/np.pi * np.ones_like(t)

for n in range(1, N+1):
    coef = (4*V*(-1)**n) / (np.pi * (1 - (2*n)**2))
    v_sf += coef * np.cos(2*n*w*t)

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.plot(t, v_sf)
plt.title("Segundo Ejercicio")
plt.xlabel("t")
plt.ylabel("v(t)")
plt.grid(True)
plt.show()
```

## Tercer Ejercicio

```python
import numpy as np
import matplotlib.pyplot as plt


V = 1
T = 2*np.pi
w = 2*np.pi / T


t = np.linspace(0, 2*T, 4000)


v = np.zeros_like(t)

for i in range(len(t)):
    tp = t[i] % T
    if tp < T/2:
        v[i] = V
    else:
        v[i] = -V

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.plot(t, v, linewidth=2)
plt.title("Tercer Ejercicio")
plt.xlabel("t")
plt.ylabel("v(t)")
plt.grid(True)
plt.show()
```

## Quinto Ejercicio

```python
import numpy as np
import matplotlib.pyplot as plt

V = 1
T = 2*np.pi
tau = T/4                
w = 2*np.pi / T

t = np.linspace(-T, T, 4000)

v = np.zeros_like(t)

for i in range(len(t)):
    tp = ((t[i] + T/2) % T) - T/2  
    if abs(tp) <= tau/2:
        v[i] = V
    else:
        v[i] = 0

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.plot(t, v, linewidth=2)
plt.title("Quinto ejercicio")
plt.xlabel("t")
plt.ylabel("v(t)")
plt.grid(True)
plt.show()
```

## Sexto Ejercicio 

```python
import numpy as np
import matplotlib.pyplot as plt

V = 1
T = 2*np.pi

t = np.linspace(-T, T, 2000)

def triangular_par_centrada(t, T, V):
    t_mod = np.mod(t, T)
    t_mod[t_mod > T/2] -= T

    return - (4*V/T) * np.abs(t_mod) + V

v = triangular_par_centrada(t, T, V)

plt.figure(figsize=(8,3))
plt.plot(t, v, linewidth=2, color='blue')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.xlim(-T/2 - 0.3, T + 0.3)
plt.ylim(-1.2*V, 1.2*V)
plt.title("Sexto ejercicio")
plt.xlabel("t")
plt.ylabel("v(t)")
plt.grid(True)
plt.show()

```
