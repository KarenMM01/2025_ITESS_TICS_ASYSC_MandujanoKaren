```python
from thinkdsp import read_wave
import matplotlib.pyplot as plt
%matplotlib inline
```

## Audio del Alumno 1 Jesús López Silva

```python
wave = read_wave('Saludo1.wav')

print("Duración del audio (segundos):", wave.duration)
print("Frecuencia de muestreo (Hz):", wave.framerate)
print("Número de muestras:", len(wave.ys))
```

## Grafica de la forma de Onda del Audio 1

```python
wave.plot()

import matplotlib.pyplot as plt
plt.grid(True, which='both')
plt.title("Forma de onda - Audio 1")
plt.show()
```

## Espectro del Audio 1

```python
spectrum = wave.make_spectrum()

spectrum.plot()

import matplotlib.pyplot as plt
plt.grid(True, which= 'both')
plt.title("Espectro de frecuencia - Audio 1")
plt.show()
```

## Audio del Alumno 2 Diana Karen Mandujano Magaña

## Grafica de la forma de Onda del Audio 2

```python
wave2.plot()

import matplotlib.pyplot as plt
plt.grid(True, which='both')
plt.title("Forma de onda - Audio 2")
plt.show()
```

## Espectro del Audio 2

```python
spectrum2 = wave2.make_spectrum()

spectrum2.plot()

import matplotlib.pyplot as plt
plt.grid(True, which='both')
plt.title("Espectro de frecuencia - Audio 2")
plt.show()
```

## Comparacion de Ondas

```python
plt.figure(figsize=(12,4))
plt.plot(wave.ts, wave.ys, label='Audio 1')
plt.plot(wave2.ts, wave2.ys, label='Audio 2', alpha=0.7)
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.title("Comparativa de la forma de onda")
plt.legend()
plt.grid(True)
plt.show()
```

## Comparacion de Espectro de Frecuencia
